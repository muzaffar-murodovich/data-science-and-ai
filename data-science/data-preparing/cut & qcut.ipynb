```python
import pandas as pd
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/anvarnarz/praktikum_datasets/main/world_population_duplicates.csv', usecols=['country','pop2021','area'], index_col='country')
df.head()
```

```python
df.loc['Uzbekistan']
```

```python
bins = [0, 1000, 10000, 30000, 50000, 100000, 300000, 1500000]
population = df.pop2021
population
```

```python
group = pd.cut(population, bins)
group
```

```python
group.value_counts()
```

```python
df['group'] = group
df.head()
```

```python
df = pd.read_csv("https://github.com/anvarnarz/praktikum_datasets/raw/main/automobile_data_nan.csv", index_col=0)
df.head()
```

```python
df.price.describe()
```

```python
bins = [0, 10000, 20000, 50000]
range = pd.cut(df.price, bins)
```

```python
range.value_counts()
```

```python
names = ["cheap", "moderate", "expensive"]
range = pd.cut(df.price, bins, labels=names)
range
```

```python
range.value_counts()
```

```python
df['range'] = range
df
```

```python
range = pd.cut(df.price, 5)
range.value_counts()
```

```python
range = pd.qcut(df.price, 5)
range.value_counts()
```
