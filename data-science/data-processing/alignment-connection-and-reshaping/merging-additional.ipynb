```python
import pandas as pd
import numpy as np
```

```python
cars_specs = pd.DataFrame({
    "car_id": np.arange(1, 6),
    "make": np.random.choice(["Toyota", "Ford", "BMW", "Honda"], 5),
    "year": np.random.randint(2015, 2023, 5),
    "engine_liters": np.round(np.random.uniform(1.4, 3.5, 5), 1)
})

cars_sales = pd.DataFrame({
    "car_id": np.arange(3, 8),
    "price_usd": np.random.randint(15000, 45000, 5),
    "mileage_km": np.random.randint(5000, 120000, 5)
})

cars_specs
```

```python
cars_sales
```

```python
merged = pd.merge(cars_specs, cars_sales)
merged
```

```python
merged = pd.merge(cars_specs, cars_sales, how='outer')
merged
```

```python
merged = pd.merge(cars_specs, cars_sales, how='left')
merged
```

```python
merged = pd.merge(cars_specs, cars_sales, left_on='car_id', right_on='car_id')
merged
```

```python
cars_specs = cars_specs.set_index('car_id')
cars_specs
```

```python
merged = pd.merge(cars_specs, cars_sales, left_index=True, right_on='car_id', how='inner')
merged
```
