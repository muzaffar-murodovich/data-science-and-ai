```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("https://github.com/anvarnarz/praktikum_datasets/raw/main/automobile_data.csv", index_col=0)
df.head()
```

```python
df.company.value_counts()
```

```python
df[df.company=='audi']
```

```python
df.groupby(['company', 'body-style']).price.max(numeric_only=True)
```

```python
df.groupby("company").agg(
          avg_price=("price", "mean"),
          max_horsepower=("horsepower", "max"),
          min_mileage=("average-mileage", "min"),
          avg_length=("length", "mean"),
          car_count=("price", "count")
      )
```

```python
df.groupby("company").agg(
    wheel_base_min=("wheel-base", "min"),
    wheel_base_max=("wheel-base", "max"),
    wheel_base_mean=("wheel-base", "mean"),
    length_min=("length", "min"),
    length_max=("length", "max"),
    length_mean=("length", "mean"),
    horsepower_min=("horsepower", "min"),
    horsepower_max=("horsepower", "max"),
    horsepower_mean=("horsepower", "mean"),
    mileage_min=("average-mileage", "min"),
    mileage_max=("average-mileage", "max"),
    mileage_mean=("average-mileage", "mean"),
    price_min=("price", "min"),
    price_max=("price", "max"),
    price_mean=("price", "mean"),
)
```
