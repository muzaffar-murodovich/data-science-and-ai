```python
import pandas as pd
import sqlite3 as sql
```

```python
db = sql.connect("newdb.db")
cursor = db.cursor()
```

```python
df = pd.read_csv("https://raw.githubusercontent.com/muzaffar-murodovich/data-science-and-ai/refs/heads/master/data-science/data-processing/xlsx2csv.csv", index_col=0)
df.head()
```

```python
new_cols = ['firstname', 'lastname', 'gender', 'country', 'age', 'date', 'id']
df.columns = new_cols
df.head()
```

```python
df.to_sql('ppl', db)
```

```python
cursor.execute("SELECT name FROM sqlite_master WHERE type='table';")
cursor.fetchall()
```

```python
cursor.execute("PRAGMA table_info(ppl);")
cursor.fetchall()
```
