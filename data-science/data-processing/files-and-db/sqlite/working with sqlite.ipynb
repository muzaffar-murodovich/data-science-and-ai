```python
import pandas as pd
import sqlite3 as sql
```

```python
db = sql.connect("chinook.sqlite")
cursor = db.cursor()
```

```python
cursor.execute("SELECT name FROM sqlite_master WHERE type='table';")
cursor.fetchall()
```

```python
cursor.execute("SELECT COUNT (*) FROM Genre")
cursor.fetchone()
```

```python
cursor.execute("PRAGMA table_info(Track);")
cursor.fetchall()
```

```python
cursor.execute("SELECT * FROM Customer;")
cursor.fetchone()
```

```python
df = pd.read_sql("SELECT * FROM Customer LIMIT 5;", db, index_col="CustomerId")
df
```

```python
df = pd.read_sql("SELECT * FROM Artist;", db, index_col="ArtistId")
df.head(10)
```

```python
df = pd.read_sql("SELECT CustomerId, FirstName, City FROM Customer LIMIT 10;", db, index_col="CustomerId")
df
```

### Ordering & sorting

```python
command = "SELECT CustomerId, FirstName, LastName, Phone, City FROM Customer ORDER BY FirstName LIMIT 10;"
df = pd.read_sql(command, db, index_col="CustomerId")
df
```

```python
command = "SELECT DISTINCT Country from Customer;"
cursor.execute(command)
cursor.fetchall()
```

### Filtering

```python
query = "SELECT * FROM Customer WHERE Country='Norway'"
pd.read_sql(query, db)
```
