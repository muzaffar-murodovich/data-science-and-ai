```python
import numpy as np
```

```python
my_list = list(range(100000)) # python list (0-99999) → normal
my_array = np.array(range(100000)) # numpy array(massive) (0-99999) → vector
```

```python
%time for _ in range(100): [x*2 for x in my_list] #normal
```

```python
%time for _ in range(100): my_array*2 #vector
```
