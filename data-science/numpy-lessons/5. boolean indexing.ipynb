```python
import numpy as np
```

```python
# Names

names = np.array(['Eve', 'Bob', 'Charlie', 'David', 'Eve', 'Frank', 'Grace'])
names
```

```python
# data
data = np.random.randn(7, 4)  # 7 rows, 4 columns
data
```

```python
names == 'Eve'  # Boolean array where names are 'Eve'
```

```python
data[names == 'Eve']  # Select rows where names are 'Eve'
```

```python
data[names != 'Eve']  # Select rows where names are not 'Eve'
```

```python
print(names)
print(data)
```

```python
mask = (names == 'Eve') | (names == 'Bob')  # Combine conditions
data[mask]  # 'Eve' or 'Bob'
```
