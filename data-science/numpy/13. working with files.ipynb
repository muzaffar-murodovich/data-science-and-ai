```python
import numpy as np
```

```python
arr = np.arange(10)
arr
```

```python
np.save('data.npy', arr)
```

```python
arr2 = np.load('data.npy')
arr2
```

```python
arr1 = np.arange(20)
np.savez('arrays', a=arr, b=arr1)
```

```python
arrays = np.load('arrays.npz')
arrays['a']
```

```python
arrays['b']
```

```python
np.savez_compressed('arrays_compressed', a=arr, b=arr1)
```

```python
c_arrays = np.load('arrays_compressed.npz')
c_arrays['a']
```
