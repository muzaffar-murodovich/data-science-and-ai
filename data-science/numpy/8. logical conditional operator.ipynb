```python
import numpy as np
```

```python
xarr = np.array([1.1, 1.2, 1.3, 1.4, 1.5])
yarr = np.array([2.1, 2.2, 2.3, 2.4, 2.5])
```

```python
# condition
cond = np.array([True, False, True, True, False])
```

```python
# where 
result = np.where(cond, xarr, yarr) #if cond is True, take from xarr else from yarr
result
```
