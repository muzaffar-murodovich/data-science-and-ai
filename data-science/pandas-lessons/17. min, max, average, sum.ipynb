### Core Descriptive Statistics
| Function | Purpose |
|----------|---------|
| df.describe() | Summary statistics (count, mean, std, min, quartiles, max) |
| df.mean() | Mean of values |
| df.median() | Median |
| df.mode() | Most frequent value(s) |
| df.std() | Standard deviation |
| df.var() | Variance |
| df.min() | Minimum |
| df.max() | Maximum |
| df.sum() | Sum of values |
| df.count() | Non-null count |


### Distribution & Shape
| Function | Purpose |
|----------|---------|
| df.quantile(q) | Percentiles / quantiles |
| df.skew() | Skewness |
| df.kurt() | Kurtosis |
| df.nunique() | Number of unique values |
| df.value_counts() | Frequency table (Series) |

### Missing Data Statistics
| Function | Purpose |
|----------|---------|
| df.isna().sum() | Count missing values per column |
| df.notna().sum() | Count non-missing values |
| df.dropna() | Remove missing values |
| df.fillna(value) | Replace missing values |


### Relationships & Comparison
| Function | Purpose |
|----------|---------|
| df.corr() | Correlation matrix |
| df.cov() | Covariance matrix |
| df.compare(other) | Compare two DataFrames |


### Group-Based Statistics
| Function | Purpose |
|----------|---------|
| df.groupby(col).mean() | Mean per group |
| df.groupby(col).sum() | Sum per group |
| df.groupby(col).agg() | Multiple statistics at once |
| df.pivot_table() | Spreadsheet-style aggregation |


### Dataset Overview
| Function | Purpose |
|----------|---------|
| df.info() | Data types, non-null counts |
| df.shape | Rows and columns |
| df.dtypes | Column data types |
| df.head() / df.tail() | Preview data |
