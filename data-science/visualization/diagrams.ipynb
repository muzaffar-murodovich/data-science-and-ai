```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("../pandas-lessons/uzb.csv", index_col=0)
df.columns = ['population', 'area']
df
```

```python
plt.figure(figsize=(16,6))
sns.barplot(x=df.index, y=df['population'])
plt.title("Population of Uzbekistan")
plt.xlabel("Area")
plt.ylabel("Population (million)")
# plt.grid()
plt.xticks(rotation=90)
plt.show()
```
