```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("../pandas-lessons/merc.csv")
df = df[df.year>=2010]
df = df[df.transmission!='Other'] 
df = df[df.fuelType!='Other']
df.head()
```

```python
df.shape
```

```python
plt.figure(figsize=(10,5))
sns.histplot(data=df, x='price')
plt.xticks(rotation=90)
plt.show()
```

```python
sns.displot(data=df, x='mileage', height=5, aspect=2, hue="transmission")
plt.xticks(rotation=90)
plt.show()
```

```python
sns.displot(data=df, x="price", col="transmission", height=5, aspect=2)
plt.show()
```

# KDE

```python
sns.displot(data=df, x='price', kind='kde', height=5, aspect=2)
plt.xticks(rotation=90)
plt.show()
```

```python
sns.displot(data=df, x='price', kde=True, height=5, aspect=2)
plt.show()
```

```python
#kdeplot
plt.figure(figsize=(10,5))
sns.kdeplot(data=df, x='price', hue='fuelType', fill=True)
plt.xticks(rotation=90)
plt.show()
```

```python
sns.displot(data=df, x='price', hue='fuelType', kind='kde', height=5, aspect=2)
plt.xticks(rotation=90)
plt.show()
```

```python
sns.displot(data=df, x="price", col="fuelType", kind='kde', height=5, aspect=1)
plt.show()
```

# CDF

```python
sns.displot(data=df, x="mileage", kind="ecdf", height=5, aspect=2)
plt.xticks(rotation=90)
plt.grid(which='both')
plt.minorticks_on()
plt.show()
```

```python
sns.displot(data=df, x="price", y="year", kind="kde", height=5, aspect=2)
plt.xticks(rotation=90)
plt.show()
```

```python
sns.displot(data=df, x="price", hue="fuelType", kind="ecdf", height=5, aspect=2)
plt.xticks(rotation=90)
plt.grid()
plt.show()
```

```python
plt.figure(figsize=(10,5))
sns.ecdfplot(data=df, x="price", hue="transmission")
plt.grid()
plt.show()
```
