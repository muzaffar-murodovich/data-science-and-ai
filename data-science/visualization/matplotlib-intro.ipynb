```python
x = np.arange(0, 4*np.pi, 0.1)
y = np.sin(x)
```

```python
plt.figure(figsize=(10, 4))
plt.plot(x, y, linewidth=3, linestyle=":", color="g", marker="o")
plt.title("func sin(x)")
plt.xlabel("pi")
plt.ylabel("sin(x)")
plt.grid()
```

```python
x = np.arange(0, 4 * np.pi, 0.1)
sin = np.sin(x) # sin(x)
cos = np.cos(x) # cos(x)

plt.figure(figsize=(12,6))
# sin(x) 
plt.plot(x, sin,
         linewidth=1.5,
         linestyle='-.',
         color='#8E44AD',
         alpha=0.7,
         marker='h')

# cos(x) 
plt.plot(x, cos,
         linewidth=1.5,
         linestyle=':',
         color='c',
         alpha=0.7,
         marker='v')

plt.title('sin(x) & cos(x)')
plt.xlabel("pi")
plt.legend(['sin(x)','cos(x)'], loc=0, fontsize='x-large')
plt.grid(color='tomato', alpha=0.5, linestyle='--')
plt.savefig('sin-cos.png')
plt.show()
```
