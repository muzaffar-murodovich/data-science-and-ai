```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("../pandas-lessons/merc.csv")
df
```

```python
df=df[df.year>2000]
plt.figure(figsize=(10,6))
# sns.scatterplot(data=df, x='year', y='price')
sns.regplot(data=df, x='mileage', y='price')
plt.show()
```
